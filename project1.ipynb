```python
n=df.replace(to_replace="Other",value="Female")
n
```

```python
#Gender Distridution
```

```python
plt.figure(figsize=(4,3))
ax=sns.countplot(data=n, x="gender")
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution')
plt.show()
```

```python
#from the above chart we have analyzed that:
#the number of females in the data is more than the number of males
```

```python
gb=df.groupby('parent_education').agg({"math_score":'mean',"reading_score":'mean',"writing_score":'mean'})
gb
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(gb,annot=True)
plt.title('Relation Between Parent Education')
plt.show()
```

```python
#from the above chart we have concluded that the education of parents have a good impact on their children
```

```python
gb1=df.groupby('internet_access').agg({"math_score":'mean',"reading_score":'mean',"writing_score":'mean'})
gb1
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(gb1,annot=True)
plt.title('Relation Between Internet Access')
plt.show()
```

```python
#from the above chart we have analyze that internet access have no impact on students marks
```

```python
nd=df.replace(to_replace = ['Free or reduced', 'Standard'], value = ['group A','group B'])
nd
```

```python
print(nd['lunch_type'].unique())
```

```python
groupA=nd.loc[(nd['lunch_type'] == "group A")].count()
groupB=nd.loc[(nd['lunch_type'] == "group B")].count()
l = ["group A", "group B"]
ml = [groupA['lunch_type'],groupB['lunch_type']] 
print(ml)
plt.title('Distribution of lunch type')
plt.pie(ml, labels=l, autopct="%1.2f%%")
plt.show()
```

```python
aq=sns.countplot(data=nd, x = "lunch_type")
aq.bar_label(aq.containers[0])
plt.title('Distribution of Lunch Type')
plt.show()
```
